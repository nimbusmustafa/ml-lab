```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data={
    'mother height':[58,62,60,64,67,70],
    'daughter height':[60,60,58,60,70,72]
}

df=pd.DataFrame(data)
df.to_csv('q.csv', index=False)
```

```python
df=pd.read_csv('q.csv')
x=df['mother height'].values
y=df['daughter height'].values
b0=0.0
b1=0.0
epochs=4
alpha=0.00004
for epoch in range(epochs):
    for i in range(x.shape[0]):
        xi=x[i]
        yi=y[i]
        yhat=(b1*xi)+b0
        error= yhat-yi
        b0=b0-alpha*error
        b1=b1-alpha*error*xi

print(b0)
print(b1)
b0,b1
```

```python
y_pred= b1*x+b0
rmse=np.sqrt(np.mean((y-y_pred)**2))
print(rmse)
```

```python
df=pd.read_csv('q.csv')
x=df['mother height'].values
y=df['daughter height'].values
x_b=np.c_[np.ones(x.shape), x]
theta=np.linalg.inv(x_b.T@x_b)@x_b.T@y
b0,b1=theta
print(b0)
print(b1)
```

```python
df=pd.read_csv('q1.csv')
X=df['hours of study'].values
Y=df['pass'].values
b0,b1=0,0
alpha=0.0001
for _ in range(500):
        for x,y in zip(X,Y):
            z=b1*x + b0
            y_pred=1/(1+np.exp(-z))
            error=y_pred-y
            print(error)
            b1-=(alpha*error*x)
            b0-=(alpha*error)

print(b0,b1)
```

```python
df=pd.read_csv('q1.csv')
x=df['hours of study'].values
y=df['pass'].values
b0,b1=0,0
a=0.01
for i in range(500):
    for j in range(y.shape[0]):
        z = b0 + (b1 * x[j])
        p = 1 / (1 + np.exp(-z))
        err = (p-y[j])
        print(err)
        b0 = b0 - (a * err )
        b1 = b1 - (a * err * x[j])




print(b0,b1)

        
```

```python
y_probs=[1/(1+np.exp(-(b1*x1+b0))) for x1 in x]
print(y_probs)
y_binary = [1 if prob > 0.5 else 0 for prob in y_probs]
print(y_binary)
```
